
## Blending Images

```python
import cv2
```

```python
# Two images
img1 = cv2.imread('flat_chessboard.png')
img2 = cv2.imread('watermark_no_copy.png')
```

```python
img1.shape
```

```python
img2.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img1)
```

Whoops! Let's remember to fix the RGB!

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

### Resizing the Images

```python
img1 =cv2.resize(img1,(1200,1200))
img2 =cv2.resize(img2,(1200,1200))
```

Let's practice resizing the image, since the DO NOT COPY image is actually quite large 1200 by 1200, and our puppy in backpack image is 1400 by 1000

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

### Blending the Image


```python
img1.shape
```

```python
img2.shape
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.3,gamma=0)
```

```python
plt.imshow(blended)
```
